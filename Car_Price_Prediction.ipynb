#### **split dataframe**

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('car_purchase_amount', axis=1)
y = df['car_purchase_amount']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mean_absolute_error = mean_absolute_error(y_test, y_pred)
mean_squared_error = mean_squared_error(y_test, y_pred)
root_mean_squared_error = np.sqrt(mean_squared_error)   
r2_score = r2_score(y_test, y_pred)
```

### **Create Pipeline**

```python
from sklearn.pipeline import Pipeline
import joblib
import sklearn
```

```python
df = df0.copy()
```

```python
df.columns = [col.replace(' ', '_') for col in df.columns]
df = df.drop(['customer_name', 'customer_e-mail', 'country'], axis=1)
```

```python
X = df.drop('car_purchase_amount', axis=1)
y = df['car_purchase_amount']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
pipeline = Pipeline([
    ('scaler', MinMaxScaler()),
    ('model', LR())
])
```

```python
pipeline.fit(X_train, y_train)
```

```python
print("pandas==" + (pd.__version__))
print("numpy==" + (np.__version__))
print("sklearn==" + (sklearn.__version__))
print("matplotlib==" + (matplotlib.__version__))
print("joblib==" + str(joblib.__version__))
```

```python
with open('Car_Price_Prediction.pkl', 'wb') as f:
    joblib.dump(pipeline, f)

print("Pipeline has been saved to Car_Price_Prediction.pkl")
```

```python
with open('Car_Price_Prediction.pkl', 'rb') as f:
    loaded_pipeline = joblib.load(f)

print("Pipeline has been loaded from Car_Price_Prediction.pkl")
```

```python
y_pred = loaded_pipeline.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error of the loaded pipeline: {mse:.2f}")
```
